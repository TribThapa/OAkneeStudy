```python
matrix = np.arange(69696).reshape(264,264)

x_value = df["NodeX"].to_numpy()
y_value = df[['NodeY']].to_numpy()
t_value = df[['Tvalue']].to_numpy()

x_axis_labels = NodeLabels.iloc[:, 3].tolist()

# creating subplot
fig, ax = plt.subplots(figsize=(20, 10))
  
# drawing heatmap on current axes
ax = sns.heatmap(matrix)
#ax.set_xticks(x_axis_labels)

for x in range(len(x_value)):
    #print(x)
    ax.add_patch(Rectangle((x_value[[x]], y_value[[x]]), 1, 1, 
                            fill=False,
                            edgecolor='blue',
                            lw=3))
plt.show()
```
